This code was used for generating the histogram use to display runtimes of the fine-tuning experiments.

```python
import numpy as np
import matplotlib.pyplot as plt

rob = [49, 28]
b560 = [150, 104]
b1b7 = [364, 242]
b3b = [701, 462]
 
# set width of bar
barWidth = 0.25
 
# set height of bar
data = np.array([
    [49,150, 364,701],
    [28, 104, 242, 462],
    [32, 124, 305, 581]
])

with plt.style.context('ggplot'):
    fig = plt.subplots(figsize =(10, 7))

    # fig.patch.set_facecolor('lightgrey')
    # fig.patch.set_alpha(0.3)

    # Set position of bar on X axis
    br1 = np.arange(len(data[0]))
    br2 = [x + barWidth for x in br1]
    br3 = [x + barWidth for x in br2]
    
    # Make the plot
    plt.bar(br1, data[0], color ='tab:blue', width = barWidth,
            edgecolor ='grey', label ='Wiki')
    plt.bar(br2, data[1], color ='tab:red', width = barWidth,
            edgecolor ='grey', label ='CRA')
    plt.bar(br3, data[2], color ='tab:orange', width = barWidth,
            edgecolor ='grey', label ='Mixed')
    
    largest = np.max(data)
    for i in range(len(data)):
        for j in range(len(data[i])):
            plt.text(j+barWidth*i, data[i,j]+largest/100, f"{data[i,j]} min", ha = 'center')


    # Adding Xticks
    plt.xlabel('x: model', fontsize = 15)
    plt.ylabel('y: runtime (minutes)', fontsize = 15)
    plt.xticks([r + barWidth*1 for r in range(4)],
            ['RoBERTa', 'Bloomz-560m', 'Bloomz-1b7', 'Bloomz-3b'])
    
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()
```
